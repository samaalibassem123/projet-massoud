```python
import pandas as pd
```

```python
df = pd.read_csv("Emploi-parSecteur-DATASET.csv")
```

```python
df.head()
```

```python
df["Secteur d''activité: Nom"].values
```

```python
df["Secteur d''activité: Nom"] = df["Secteur d''activité: Nom"].apply(lambda x: 'Education Santé et services administratifs' if x == 'Education, Santé et services administratifs' else x)
```

```python
df.head()
```

```python
df = df[~df["Secteur d''activité: Nom"].str.contains(
    "Population occupée agée de 15 ans et plus",
    case=False, na=False
)]


df = df[~df["Secteur d''activité: Nom"].str.contains(
    "Non Declaré",
    case=False, na=False
)]
```

```python
df.to_csv('Fin-Emploi-parSecteur-DATASET.csv')
```
